```python
from ultralytics import YOLO
model = YOLO("yolov8n.pt")
model.train(data=r"C:\Users\JAY\Downloads\Vehicle Detection 1.v1i.yolov8\data.yaml", epochs=10, imgsz=640)
```

```python
metrics = model.val()
print(metrics)
```

```python
results = model.predict(
    source=r"C:\Users\JAY\Downloads\Vehicle Detection 1.v1i.yolov8\test\images",
    save=True,
    imgsz=640,
    conf=0.25
)
```
